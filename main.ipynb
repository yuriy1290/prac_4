```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.read_csv('vgsales_1.csv')
df
```

```python
dff = df.drop(columns=['Genre'])
dff.head(10)
```

```python
df.isnull().sum()
```

```python
dfz = df.replace({'Publisher' : np.nan}, 'ОАО ОАО')
dfz.isnull().sum()
```

```python
dfq = df
simple = SimpleImputer(strategy='median', missing_values=np.nan)
simple = simple.fit(dfq[['Year']])
dfq['Year'] = simple.transform(dfq[['Year']])
dfq.isnull().sum()
```

```python
df.sort_values(by='Name').head(10)
```

```python
df.to_csv('df.csv')
dff.to_csv('dff.csv')
dfq.to_csv('dfq.csv')
dfz.to_csv('dfz.csv')
```
